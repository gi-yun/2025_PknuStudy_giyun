```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

import pandas as pd
from datetime import datetime
```

```python
# font  D2coding
d2_path = "./ref/D2Coding-Ver1.3.2-20180524.ttf"
fm.fontManager.addfont(d2_path)
matplotlib.rcParams["font.family"] = "D2Coding"
```

```python
file_path = "./ref/_______.csv"
bir= pd.read_csv(file_path, encoding="EUC-KR")
# bir.columns = [
#     "Birth_Num",
#     "Death_Num",
#     "Ntrl_Incr_Num",
#     "Crd_Birth_Rate",
#     "Crd_Death_Rate",
#     "Ntrl_Thsd_Rate",
#     "Total_Birth_Rate",
#     "Birth_Sex_Ratio",
#     "Infant_Mrtlt_Rate",
#     #?????            ? ..?
# ]
bir = bir.loc[:,:'()']
bir.columns = [
    "date",
    "Birth_Num",
    "Death_Num",
    "Ntrl_Incr_Num",
    "Crd_Birth_Rate",
    "Crd_Death_Rate",
    "Ntrl_Thsd_Rate",
    "Total_Birth_Rate",
] #   ...
bir.tail() 
```

```python
bir['date'] = bir['date'].map(lambda x: x.rstrip(' p)')) # p) 
bir
```

```python
ex1_data = bir.copy()
r=[]
for i in range(1970,2025,1):
    r.append(str(i))
year =r
ex1_data = ex1_data[ex1_data['date'].isin(year)]
ex1_data.sample(10)
```

```python
fig , ax = plt.subplots(figsize=(15,5))

date = year
for d in date :
    rate_data = ex1_data[ex1_data['date']==d]
    x_data = rate_data['Birth_Num']
    y_data = rate_data['Death_Num']
    ax.scatter( x_data , y_data, s = 20, label = d )

ax.set_xlabel("")
ax.set_ylabel("")
ax.legend(loc="best")
ax.plot([0,1500000],[0,700000])

plt.show()
    
```

```python
fig, ax = plt.subplots(figsize=(15,5))

rate = year
for r in rate:
    rate_data = ex1_data[ex1_data['date'] == r]
    x_data = rate_data['Birth_Num']
    y_data = rate_data['Death_Num']
    
    # 
    ax.scatter(x_data, y_data, s=20, label=f"Scatter {r}")
    
    #   (    )
    ax.plot(x_data, y_data, linestyle='-', marker='o', label=f"Line {r}")

ax.set_xlabel("")
ax.set_ylabel("")

plt.show()
```
