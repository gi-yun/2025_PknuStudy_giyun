```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

import pandas as pd
from datetime import datetime
```

```python
# font 설정 D2coding
d2_path = "./ref/D2Coding-Ver1.3.2-20180524.ttf"
fm.fontManager.addfont(d2_path)
matplotlib.rcParams["font.family"] = "D2Coding"
```

```python
file_path = "./ref/인구동태건수_및_동태율_추이_출생_사망_혼인_이혼.csv"
bir= pd.read_csv(file_path, encoding="EUC-KR")
# bir.columns = [
#     "Birth_Num",
#     "Death_Num",
#     "Ntrl_Incr_Num",
#     "Crd_Birth_Rate",
#     "Crd_Death_Rate",
#     "Ntrl_Thsd_Rate",
#     "Total_Birth_Rate",
#     "Birth_Sex_Ratio",
#     "Infant_Mrtlt_Rate",
#     #????? 생각해보니까 이런 뺑이 칠필요없이 그냥 몇개 걸러낸다음에 다시 이름지어주는게 더 빠르지 않을까? 어..?
# ]
bir = bir.loc[:,:'합계출산율(명)']
bir.columns = [
    "date",
    "Birth_Num",
    "Death_Num",
    "Ntrl_Incr_Num",
    "Crd_Birth_Rate",
    "Crd_Death_Rate",
    "Ntrl_Thsd_Rate",
    "Total_Birth_Rate",
] # 왜 뻘짓하고 있었지...
bir.tail() 
```

```python
bir['date'] = bir['date'].map(lambda x: x.rstrip(' p)')) # p) 제거
bir
```

```python
ex1_data = bir.copy()
r=[]
for i in range(1970,2025,1):
    r.append(str(i))
year =r
ex1_data = ex1_data[ex1_data['date'].isin(year)]
ex1_data.sample(10)
```

```python
fig , ax = plt.subplots(figsize=(15,5))

date = year
for d in date :
    rate_data = ex1_data[ex1_data['date']==d]
    x_data = rate_data['Birth_Num']
    y_data = rate_data['Death_Num']
    ax.scatter( x_data , y_data, s = 20, label = d )

ax.set_xlabel("출생")
ax.set_ylabel("사망")
ax.legend(loc="best")
ax.plot([0,1500000],[0,700000])

plt.show()
    
```

```python
fig, ax = plt.subplots(figsize=(15,5))

rate = year
for r in rate:
    rate_data = ex1_data[ex1_data['date'] == r]
    x_data = rate_data['Birth_Num']
    y_data = rate_data['Death_Num']
    
    # 산점도
    ax.scatter(x_data, y_data, s=20, label=f"Scatter {r}")
    
    # 선 그래프 (출생과 사망 데이터 순서대로 연결)
    ax.plot(x_data, y_data, linestyle='-', marker='o', label=f"Line {r}")

ax.set_xlabel("출생")
ax.set_ylabel("사망")

plt.show()
```
