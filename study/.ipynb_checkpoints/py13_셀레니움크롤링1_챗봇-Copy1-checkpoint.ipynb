```python
import os
import json
import datetime
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
start = time.time()

CACHE_FILE = "hanmirak_cache.json"

# âœ… ìºì‹œ ì½ê¸°
def read_cache():
    if os.path.exists(CACHE_FILE):
        with open(CACHE_FILE, "r", encoding="utf-8") as f:
            try:
                return json.load(f)
            except:
                return {}
    return {}

# âœ… ìºì‹œ ì €ì¥
def write_cache(data):
    with open(CACHE_FILE, "w", encoding="utf-8") as f:
        json.dump(data, f, ensure_ascii=False)

# âœ… í¬ë¡¤ë§

def _fetch_hanmirak_table():
    options = Options()
    options.binary_location = "./chrome-win64/chrome.exe"
    options.add_argument("--headless")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("--window-size=1920,1080")
    options.add_argument("--disable-gpu")
    options.add_argument("--log-level=3")
    prefs = {
        "profile.managed_default_content_settings.images": 2,
        "profile.managed_default_content_settings.fonts": 2,
        "profile.managed_default_content_settings.plugins": 2
    }
    options.add_experimental_option("prefs", prefs)

    service = Service("chromedriver.exe")
    driver = webdriver.Chrome(service=service, options=options)
    driver.get("https://www.pknu.ac.kr/main/399")

    try:
        wait = WebDriverWait(driver, 10)
        link = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, "#subCont > table > tbody > tr:nth-child(1) > td.bdlTitle > a")))
        ActionChains(driver).move_to_element(link).click().perform()
        wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "div.con03_sub_2_wrap")))
        soup = BeautifulSoup(driver.page_source, "html.parser")

        for div in soup.select("div.con03_sub_2_wrap"):
            p = div.find("p")
            if p and "í•œë¯¸ë½" in p.text:
                return div.find("table", class_="con03_sub_2")
        return None
    finally:
        driver.quit()

# âœ… ìºì‹œ ì´ˆê¸°í™” (ìµœì´ˆ ì‹¤í–‰ ì‹œ ì¼ì£¼ì¼ì¹˜ ìºì‹±)
def init_weekly_cache():
    table = _fetch_hanmirak_table()
    if not table:
        return "[ì˜¤ë¥˜] í•œë¯¸ë½ ì‹ë‹¨ ì •ë³´ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤."

    days = ['ì›”', 'í™”', 'ìˆ˜', 'ëª©', 'ê¸ˆ']
    meal_row = table.select('tr')[2]
    tds = meal_row.select('td')[1:6]

    weekly_cache = {}
    for idx, td in enumerate(tds):
        meals = [p.get_text(strip=True) for p in td.select('p') if p.get_text(strip=True)]
        content = "\n".join(f"\u2022 {m}" for m in meals) if meals else "\u2022 ì‹ë‹¨ ì •ë³´ ì—†ìŒ"
        weekly_cache[days[idx]] = content

    write_cache(weekly_cache)
    return weekly_cache

# âœ… ì˜¤ëŠ˜ ì‹ë‹¨ ë³´ê¸°

def get_today_menu():
    weekday = datetime.datetime.today().weekday()
    if weekday >= 5:
        return "ğŸ˜´ ì˜¤ëŠ˜ì€ ì£¼ë§ì´ë¼ í•œë¯¸ë½ ìš´ì˜ì´ ì—†ìŠµë‹ˆë‹¤."

    days = ['ì›”', 'í™”', 'ìˆ˜', 'ëª©', 'ê¸ˆ']
    cache = read_cache()
    if not cache:
        cache = init_weekly_cache()

    return f"ğŸ“… ì˜¤ëŠ˜ì˜ í•œë¯¸ë½ ì‹ë‹¨ ({days[weekday]})\n{cache.get(days[weekday], 'â€¢ ì‹ë‹¨ ì •ë³´ ì—†ìŒ')}"

# âœ… ì£¼ê°„ ì‹ë‹¨ ë³´ê¸°

def get_weekly_menu():
    cache = read_cache()
    if not cache:
        cache = init_weekly_cache()

    return "\n\n".join([f"ğŸ“… {day}ìš”ì¼\n{cache.get(day, 'â€¢ ì‹ë‹¨ ì •ë³´ ì—†ìŒ')}" for day in ['ì›”', 'í™”', 'ìˆ˜', 'ëª©', 'ê¸ˆ']])
print(get_today_menu())
print("ê±¸ë¦° ì‹œê°„:", time.time() - start)
print(get_weekly_menu())
print("ê±¸ë¦° ì‹œê°„:", time.time() - start)
```

```python
import os
import json
import datetime
import time
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# ====== ìºì‹œ ê²½ë¡œ ë° íŒŒì¼ ======
CACHE_FILE = "hanmirak_cache.json"

def read_cache():
    if os.path.exists(CACHE_FILE):
        with open(CACHE_FILE, "r", encoding="utf-8") as f:
            try:
                return json.load(f)
            except:
                return {}
    return {}

def write_cache(data):
    with open(CACHE_FILE, "w", encoding="utf-8") as f:
        json.dump(data, f, ensure_ascii=False, indent=2)

# ====== í•œë¯¸ë½ í…Œì´ë¸” HTML ê°€ì ¸ì˜¤ê¸° ======
def _fetch_hanmirak_table():
    options = Options()
    options.binary_location = "./chrome-win64/chrome.exe"
    # options.add_argument("--headless")
    options.add_argument("--no-sandbox")
    options.add_argument("--disable-dev-shm-usage")
    options.add_argument("--window-size=1920,1080")
    options.add_argument("--disable-gpu")
    options.add_argument("--log-level=3")
    prefs = {
        "profile.managed_default_content_settings.images": 2,
        "profile.managed_default_content_settings.fonts": 2,
        "profile.managed_default_content_settings.plugins": 2
    }
    options.add_experimental_option("prefs", prefs)

    service = Service("chromedriver.exe")
    driver = webdriver.Chrome(service=service, options=options)
    driver.get("https://www.pknu.ac.kr/main/399")

    try:
        wait = WebDriverWait(driver, 10)
        link = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, "#subCont > table > tbody > tr:nth-child(1) > td.bdlTitle > a")))
        ActionChains(driver).move_to_element(link).click().perform()

        wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, "div.con03_sub_2_wrap")))
        soup = BeautifulSoup(driver.page_source, "html.parser")

        for div in soup.select("div.con03_sub_2_wrap"):
            p = div.find("p")
            if p and "í•œë¯¸ë½" in p.text:
                return div.find("table", class_="con03_sub_2")
        return None
    finally:
        driver.quit()

# ====== ì „ì²´ ì£¼ê°„ ì‹ë‹¨ ìºì‹± ======
def update_cache():
    table = _fetch_hanmirak_table()
    if not table:
        return None

    cache = {}
    try:
        meal_row = table.select('tr')[2]
        tds = meal_row.select('td')[1:6]
        for idx, td in enumerate(tds):
            date = (datetime.datetime.today() + datetime.timedelta(days=(idx - datetime.datetime.today().weekday()))).strftime("%mì›” %dì¼")
            meals = [p.get_text(strip=True) for p in td.select('p') if p.get_text(strip=True)]
            cache[f"{['ì›”','í™”','ìˆ˜','ëª©','ê¸ˆ'][idx]}"] = f"{date} " + "\n" + "\n".join(f"â€¢ {m}" for m in meals) if meals else "â€¢ ì‹ë‹¨ ì •ë³´ ì—†ìŒ"
        write_cache(cache)
        return cache
    except Exception as e:
        print(f"[ë°ì´í„° ì²˜ë¦¬ ì˜¤ë¥˜] {e}")
        return None

# ====== ì˜¤ëŠ˜ ì‹ë‹¨ ê°€ì ¸ì˜¤ê¸° ======
def get_today_menu(force=False):
    weekday = datetime.datetime.today().weekday()
    days = ['ì›”','í™”','ìˆ˜','ëª©','ê¸ˆ']

    if weekday >= 5:
        return "ğŸ˜´ ì£¼ë§ì€ í•œë¯¸ë½ ìš´ì˜ì´ ì—†ìŠµë‹ˆë‹¤."

    cache = read_cache()
    if not cache or force:
        cache = update_cache()
        if not cache:
            return "[ì˜¤ë¥˜] ì˜¤ëŠ˜ì˜ ì‹ë‹¨ ì •ë³´ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤."

    return f"ğŸ“… ì˜¤ëŠ˜ì˜ í•œë¯¸ë½ ì‹ë‹¨ ({cache[days[weekday]].splitlines()[0]})\n" + "\n".join(cache[days[weekday]].splitlines()[1:])

# ====== ì´ë²ˆ ì£¼ ì‹ë‹¨ ê°€ì ¸ì˜¤ê¸° ======
def get_weekly_menu(force=False):
    days = ['ì›”','í™”','ìˆ˜','ëª©','ê¸ˆ']
    cache = read_cache()
    if not cache or force:
        cache = update_cache()
        if not cache:
            return "[ì˜¤ë¥˜] ì£¼ê°„ ì‹ë‹¨ ì •ë³´ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤."

    result = []
    for day in days:
        lines = cache.get(day, "â€¢ ì‹ë‹¨ ì •ë³´ ì—†ìŒ").splitlines()
        if lines:
            header = f"ğŸ“… {day}ìš”ì¼ ({lines[0]})"
            result.append(header + "\n" + "\n".join(lines[1:]))
        else:
            result.append(f"ğŸ“… {day}ìš”ì¼\nâ€¢ ì‹ë‹¨ ì •ë³´ ì—†ìŒ")
    return "\n\n".join(result)

# ====== ì‹¤í–‰ ======
if __name__ == "__main__":
    start = time.time()
    print("ğŸš€ í•œë¯¸ë½ ì‹ë‹¨ ì •ë³´ ê°€ì ¸ì˜¤ê¸° ì‹œì‘...\n")

    print("âœ… ì˜¤ëŠ˜ì˜ ì‹ë‹¨")
    print(get_today_menu())
    print()

    print("ğŸ“† ì´ë²ˆ ì£¼ ì‹ë‹¨")
    print(get_weekly_menu())

    print("\nâ±ï¸ ì´ ì†Œìš” ì‹œê°„:", round(time.time() - start, 2), "ì´ˆ")
```
